```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
from pylab import *
```

```python
#Using Matlab type
```

```python
x = np.linspace(0, 5, 10)
y = x ** 2
```

```python
figure()
plot(x, y, 'g^')
xlabel('x')
ylabel('y')
title('Numero Uno')
show()
```

```python
subplot(1,2,1)
plot(x, y, 'r--')
subplot(1,2,2)
plot(y, x, 'b*-');
```

```python
#Using matlab-like API
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #range 0 to 1 ?????

axes.plot(x, y, 'g--')

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Better One');
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y, 'r*')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('y=x^2 function')

axes2.plot(y, x, 'yo')
axes2.set_xlabel('x')
axes2.set_ylabel('y')
axes2.set_title('inverse function');
```
